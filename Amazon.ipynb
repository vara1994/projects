```python
from bs4 import BeautifulSoup
import requests
import urllib.request
from time import sleep
from random import randint
import numpy as np
import pandas as pd
```

```python
url='https://www.amazon.in/s?k=mobiles+under+20000+rupees&crid=25CGDL3GJDMKK&sprefix=mobiles+under+200%2Caps%2C291&ref=nb_sb_ss_ts-doa-p_1_17'
```

```python
page = requests.get(url)
type(page)
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
type(soup)
```

```python
Product_name=[]
price=[]
img_url=[]
Rating=[]
```

```python
title=soup.find_all('div',class_='a-section a-spacing-medium')
print(title)
```

```python
for details in title:
    product= details.find_all('img')
    img_url.append(product)
    print(img_url[:10])
    
    name= details.find('span', class_='a-size-medium a-color-base a-text-normal').text
    Product_name.append(name)
    print(Product_name[:10])
    
    rate=details.find('span',class_='a-price-whole').text
    price.append(rate)
    print(price[:10])
    
    rating= details.find('span', class_='a-icon-alt').text
    Rating.append(rating)
    print(Rating[:10])
    
    
```

```python
print('Mobile phones under 20000')
print(Product_name)

      
```
