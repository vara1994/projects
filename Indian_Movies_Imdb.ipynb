```python
from bs4 import BeautifulSoup
import requests
import urllib.request
from time import sleep
from random import randint
import numpy as np
import pandas as pd
```

```python
page = requests.get('https://www.imdb.com/list/ls009997493/',headers = headers)
type(page) 
```

```python
soup = BeautifulSoup(page.text, 'html.parser')
type(soup)
```

```python
headers = {'Accept-Language': 'en-US, en;q=0.5'}
```

```python
Movies_div=soup.find_all('div', class_='lister-item-content')
print(Movies_div)
```

```python
title=[]
year=[]
imdb_ratings=[]
```

```python
for container in Movies_div:
        # Scraping the movie's name
        name = container.h3.a.text
        title.append(name)
       
       
        # Scraping the movie's year
   
        years = container.h3.find('span', class_='lister-item-year').text
        year.append(years)
        
        
        
```

```python
print(title[:100])
print(year[:100])
print(imdb_ratings[:100])
```

```python
Rating= soup.find_all('div', class_='ipl-rating-widget')
type(Rating)
```

```python
for rate in Rating:
    imdb= float(rate.find('span', class_='ipl-rating-star__rating').text)
    imdb_ratings.append(imdb)
    
    
```

```python
Ind_Movies=pd.DataFrame({"Movie_Name":title[0:101], "Year":year[0:101], "Rating":imdb_ratings[0:101]})

Ind_Movies.head()
```
