```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
```

```python
input_data=pd.read_csv('sonar.csv')
print(input_data)
```

```python
print(input_data.shape)
```

```python
set_option('display.max_rows', 100)
print(input_data.dtypes)
```

```python
set_option('display.width', 100)
print(input_data.head(20))
```

```python
# descriptions, change precision to 3 places
set_option('precision', 3)
print(input_data.describe())
```

```python
fig_size = pyplot.rcParams["figure.figsize"]
fig_size[0] = 16
fig_size[1] = 12
pyplot.rcParams["figure.figsize"] = fig_size
```

```python
# Histograms for each attribute
input_data.hist()
pyplot.show()
```

```python
input_data.plot.box(figsize=(12,7), xticks=[])
plt.title('Boxplots of all frequency bins')
plt.xlabel('Frequency bin')
plt.ylabel('Power spectral density (normalized)')
plt.show()
```

```python
# Density plot for each attribute
input_data.plot(kind='density', subplots=True, layout=(8,8), sharex=False, legend=False)
pyplot.show()
```

```python
# Correlation matrix
fig = pyplot.figure()
ax = fig.add_subplot(111)
cax = ax.matshow(input_data.corr(), vmin=-1, vmax=1, interpolation='none')
fig.colorbar(cax)
pyplot.show()
```
